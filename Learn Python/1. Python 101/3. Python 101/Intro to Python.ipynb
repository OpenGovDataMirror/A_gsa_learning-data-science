## What's Python

[Python](https://www.python.org) is a **high-level**, **general purpose** programming langauge developed in 1989 by Dutch programmer [Guido van Rossum](https://en.wikipedia.org/wiki/Guido_van_Rossum). The language is named after [Monty Python's Flying Circus](https://en.wikipedia.org/wiki/Monty_Python%27s_Flying_Circus). It supports multiple programming paradigms, including object-oriented, imperative, functional and procedural, and has a large and comprehensive [standard library](https://docs.python.org/3/library/index.html).

#### High-Level
This means the languauge is strongly abstracted from the details of your computer's machine code, which is the series of 1s and 0s that tell your computer what to do. This means Python uses natural language elements, which makes code simpler and more understandable. To see this in action, compare the following Python and C codeblocks, both of which perform string concatenation to print 'hello world'.

As you can see, the Python code is much easier on the eyes. That's because Python implements a lot of the nitty gritty details behind the scenes. This allows you to produce working code much faster.

#### General Purpose
This means Python can be used just about anywhere. Being [portable](https://en.wikipedia.org/wiki/Software_portability), Python can run on many Unix variants, on the Mac, and on PCs under MS-DOS, Windows, Windows NT, and OS/2. Python can even be used as an extension language for applications that need a programmable interface.

Being general purpose also means the core Python language is quite small. This makes the language easy to learn and maintain.

This also means that you almost always need to **import** addional functionality - called **packages** - when you want to do something fun.

> **Packages** are libraries of code that are organized into files and folders, much like the files and folders on your desktop:

> A package contains:
 - files with Python code (called *modules*)
 - often some compiled code (e.g. C or FORTRAN) that can be accessed by Python
 - a file called `__init__.py` that specifies what will be executed when you import the package

> Here's an example of importing the `numpy` package, which is the fundamental package for scientific computing in Python.

```python
import numpy
```

> Once the package is imported, you have access to all of the modules within. But you're probably wondering where on Earth this `numpy` package came from. If you execute the following code, you can see where on your computer this package lives.

```python
numpy.__file__
```

> With the numpy package imported, you can now access its functionality, for example the square root operator:

```python
numpy.sqrt(16) #take the square root of 16
```

> If you know the exact functionality that you want from a package, you can import that **subpackage**. Subpackages are packages that are subdirectories of another package. Consider our square root example above. That's a subpackage. That means you could have also done this to import and execute the code:

```python
from numpy import sqrt
```

> Above, you're telling python to look inside the `numpy` directory for the subpackage called `sqrt`. If you import like that, you won't need to type `numpy.sqrt()` and can instead just type `sqrt()`:

```python
sqrt(16)
```

Python ships with a large [standard library](https://docs.python.org/3/library/index.html). If you get Python through something like Anaconda, you'll have access to even more packages.

If you're using Python through Anaconda, you can read more about installing and managing packages [here](https://conda.io/docs/user-guide/tasks/manage-pkgs.html).

### Fitting Noisy Data

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(12)

x = np.linspace(0, 1, 20)
y = np.cos(x) + 0.3*np.random.rand(20)
p = np.poly1d(np.polyfit(x, y, 3))

t = np.linspace(0, 1, 200)
plt.plot(x, y, 'o', t, p(t), '-')
plt.show()
```

### 3D Plots

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
X = np.arange(-4, 4, 0.25)
Y = np.arange(-4, 4, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='hot')
plt.show()
```

### Optimization
The six-hump camelback function 
> $f(X)=4x^2_1-2.1x^4_1+\frac{1}{3}x^6_1+x_1x_2-4x^2_2+4x^4_2$

has multiple global and local minima. We're tasked with finding the global minima of this function.

Here's a 3D surface plot of the function

```python
# Define the function that we are interested in
def sixhump(x):
    return ((4 - 2.1*x[0]**2 + x[0]**4 / 3.) * x[0]**2 + x[0] * x[1]
            + (-4 + 4*x[1]**2) * x[1] **2)

# Make a grid to evaluate the function (for plotting)
x = np.linspace(-2, 2)
y = np.linspace(-1, 1)
xg, yg = np.meshgrid(x, y)

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(xg, yg, sixhump([xg, yg]), rstride=1, cstride=1,
                       cmap=plt.cm.jet, linewidth=0, antialiased=False)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
ax.set_title('Six-hump Camelback function')
plt.show()
```

To minimize over several variables, the trick is to turn them into a function of a multi-dimensional variable (a vector):

```python
from scipy import optimize

x_min = optimize.minimize(sixhump, x0=[0, 0])

plt.figure()
# Show the function in 2D
plt.imshow(sixhump([xg, yg]), extent=[-2, 2, -1, 1])
plt.colorbar()
# And the minimum that we've found:
plt.scatter(x_min.x[0], x_min.x[1])

plt.show()
```
